```
import os
import pprint
import googleapiclient.discovery

def main():
    os.environ["OAUTHLIB_INSECURE_TRANSPORT"] = "1"

    api_service_name = "youtube"
    api_version = "v3"
    DEVELOPER_KEY = "AIzaSyAoP0YRiRCl36rqszMjq2FtVU_dXsdWWrA"

    youtube = googleapiclient.discovery.build(
        api_service_name, api_version, developerKey = DEVELOPER_KEY)

    request = youtube.commentThreads().list(
        part="snippet",
        order="relevance",
        videoId="6wXZ_m3SbEs",
        maxResults=100    )
    response = request.execute()

    pprint.pprint(response)
    cmnt=response
    return cmnt

if __name__ == "__main__":
    cmnt=main()



```

```

comments_list=[]
for i in range(len(cmnt['items'])):
  comments_list.append(cmnt['items'][i]['snippet']['topLevelComment']['snippet']['textOriginal'])
```

```
import demoji     #importing demoji used for removing emoji from text
```

```
demoji.download_codes()
```

```
comments_list_with_emoji_desc=[]                               #replacing emojis with description and storing the comments into a list
for i in comments_list:
  comments_list_with_emoji_desc.append(demoji.replace_with_desc(i , ""))
```

```
comments_list_with_emoji_desc
```

```
comments_list_with_emoji_removal=[]                               #replacing emojis with empty string and storing the comments into a list
for i in comments_list:
  comments_list_with_emoji_removal.append(demoji.replace(i, ""))
```

```
comments_list_with_emoji_removal
```

```
for i in range(len(comments_list_with_emoji_removal)):                                               #removing '\n'(i.e for next line in comments)
  comments_list_with_emoji_removal[i]=comments_list_with_emoji_removal[i].replace('\n','')
```

```
from langdetect import detect     #to segregate english comments from the given comments with landetect lib whichi 91% acc approx
```

```
english=[]
others=[]
for i in comments_list_with_emoji_removal:
  lan= detect(i)
  if (lan=="en"):
    english.append(i)
  else:
    others.append(i) 
```
